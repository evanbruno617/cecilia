```python
 import requests
from datetime import datetime, timedelta

def get_all_game_ids(start_year, end_year):
    base_url = "https://statsapi.mlb.com/api/v1/schedule"
    
    game_list = []

    while start_year != end_year:

        params = {
            "sportId": 1,  # 1 is for MLB
            "startDate": f"{start_year}-04-01",
            "endDate": f"{start_year}-09-30",
        }

        response = requests.get(base_url, params=params)
        data = response.json()

        games = data.get("dates", [])

        for game_day in games:
            date = game_day["date"]
            for game in game_day['games']:

                try:
                    game_list.append({
                    "gameId": game["gamePk"],
                    "gameDate": date,
                    "homeTeam": game["teams"]["home"]["team"]["name"],
                    "awayTeam": game["teams"]["away"]["team"]["name"],
                    "awayTeamID": game['teams']['away']['team']['id'],
                    "homeTeamID": game['teams']['home']['team']['id'],})
                except:

                    game_list.append({
                    "gameId": game["gamePk"],
                    "gameDate": date,
                    "homeTeam": np.nan,
                    "awayTeam": np.nan,
                    "awayTeamID": game['teams']['away']['team']['id'],
                    "homeTeamID": game['teams']['home']['team']['id'],
                })
        start_year += 1  


    return game_list


# Example: Get all game IDs and dates for the 2023 MLB regular season
start_year = 2010
end_year = 2025

game_list = get_all_game_ids(start_year, end_year)

# Print the results
for game in game_list[:10]:  # Print the first 10 games as a sample
    print(game)
```

```python
events_dict = []

year = 2013

for row, i in game_df.iterrows():
    
    game_id = i['gameId']
    
    date = i['gameDate']
    
    url = f"https://statsapi.mlb.com/api/v1/game/{game_id}/playByPlay"
    response = requests.get(url)
    data = response.json()


    for atbat in data['allPlays']:


        for event in atbat['playEvents']:

            if event['isPitch'] == False:
                continue
                
            try:

                occurence = atbat['result']['event']
                
            except:
                occurence= None

            try:
                
                pitch = event['details']['type']['description']
                
            except:
                
                pitch = np.nan
                
            try:
                
                location = event['hitData']['location']
                
            except:
                
                location = np.nan
                
            isout = atbat['result']['isOut']

            events_dict.append({
                'Inning': atbat['about']['inning'],
                'awayScore': atbat['result']['awayScore'],
                'homeScore': atbat['result']['homeScore'],
                'isTopInning': atbat['about']['isTopInning'],
                'Outs': atbat['count']['outs'],
                'Balls': event['count']['balls'],
                'Strikes': event['count']['strikes'],
                'X': event['pitchData']['coordinates'].get('pX', np.nan),
                'Y': event['pitchData']['coordinates'].get('pZ', np.nan),
                'Zone': event['pitchData']['coordinates'].get('zone', np.nan),
                'strikeZoneTop': event['pitchData']['strikeZoneTop'],
                'strikeZoneBottom': event['pitchData']['strikeZoneBottom'],
                'IsOut': isout,
                'Outcome': event['details']['code'],
                'Inplay': event['details']['isInPlay'],
                'Event': occurence,
                'Pitch': pitch,
                'Description': event['details']['description'],
                'Start_speed': event['pitchData'].get('startSpeed', np.nan),
                'End_speed': event['pitchData'].get('endSpeed', np.nan),
                'breakAngle': event['pitchData']['breaks'].get('breakAngle', np.nan),
                'breakLength': event['pitchData']['breaks'].get('breakLength', np.nan),
                'breakY': event['pitchData']['breaks'].get('breakY', np.nan),
                'breakVertical': event['pitchData']['breaks'].get('breakVertical', np.nan),
                'breakHorizontal': event['pitchData']['breaks'].get('breakHorizontal', np.nan),
                'breakVerticalInduced': event['pitchData']['breaks'].get('breakVerticalInduced', np.nan),
                'spinRate': event['pitchData']['breaks'].get('spinRate', np.nan),
                'location': location,
                'batter_name' : atbat['matchup']['batter']['fullName'],

                'batter_side' : atbat['matchup']['batSide']['code'],

                'batter_split' : atbat['matchup']['splits']['batter'],

                'pitcher_id' : atbat['matchup']['pitcher']['id'],

                'pitcher_name' : atbat['matchup']['pitcher']['fullName'],

                'pitcher_hand' : atbat['matchup']['pitchHand']['code'],

                'pitcher_split' : atbat['matchup']['splits']['pitcher'],
                'Date' : date
            })
            
    if i['gameDate'].split('-')[0] != year:
        
        events_df = pd.DataFrame(events_dict)
        
        events_df.to_csv(f'play_by_play/{year}.csv')
        
        year = i['gameDate'].split('-')[0]
        
        events_dict = []

```
