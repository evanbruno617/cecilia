```python
# Specify the path for the folder
folder_path = "path/to/your/folder"

# Check if the folder exists; if not, create it
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
```

```python
total_dict = {}

for file in csv_files:
    
    year = file.split('.')[0]
    
    data_df = pd.read_csv(f'play_by_play/{file}')
    
    
    batters = data_df['batter_name'].unique()
    
    for batter in batters:
        batter_df = data_df.loc[data_df['batter_name'] == batter]
        
        folder_path = f'Players/batters/{batter}'
        # Check if the folder exists; if not, create it
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)
            
        batter_df.to_csv(f'{folder_path}/{year}.csv')
        
    pitchers = data_df['pitcher_name'].unique()
    
    for pitcher in pitchers:
        pitcher_df = data_df.loc[data_df['pitcher_name'] == pitcher]
        
        folder_path = f'Players/pitchers/{pitcher}'
        # Check if the folder exists; if not, create it
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)
            
        pitcher_df.to_csv(f'{folder_path}/{year}.csv')
```

```python
total_df = pd.DataFrame()

for file in csv_files:
    
    year = file.split('.')[0]
    
    data_df = pd.read_csv(f'play_by_play/{file}')
    
    
    batters = data_df['batter_name'].unique()
    
    for batter in batters:
        batter_df = data_df.loc[data_df['batter_name'] == batter]
        
        total_df = pd.concat([total_df, batter_df])
        
```
